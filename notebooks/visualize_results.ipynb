```python
import matplotlib.pyplot as plt
import numpy as np

# Model performance data
models = ['TCN', 'PatchTST', 'ARIMA', 'TimeCMA']
mse_values = [1.434791088104248, 1.1237211227416992, 1.7947, 1.1298346636339363]
mae_values = [0.7879840731620789, 0.7177075147628784, 0.7842, 0.7263656142872726]

# Create figure with subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Color scheme - highlight best performer (PatchTST) in red, others in blue tone
base_color = '#4E95D9'  # Your specified blue
highlight_color = '#E74C3C'  # Red for best performer

# Best performer is PatchTST (index 1 - lowest MSE and 2nd lowest MAE)
colors_mse = [base_color if i != 1 else highlight_color for i in range(len(models))]
colors_mae = [base_color if i != 1 else highlight_color for i in range(len(models))]

# Set figure background
fig.patch.set_facecolor('white')

# MSE Comparison
bars1 = ax1.bar(models, mse_values, color=colors_mse, alpha=0.9, width=0.6)
ax1.set_title('Mean Squared Error (MSE)', fontsize=14, fontweight='bold', pad=25, color='#2C3E50')
ax1.set_ylabel('Score', fontsize=12, color='#2C3E50')

# Add value labels on bars
for i, (bar, value) in enumerate(zip(bars1, mse_values)):
    height = bar.get_height()
    color = 'white' if i == 1 else '#2C3E50'  # White text on red bar, dark on blue bars
    ax1.text(bar.get_x() + bar.get_width()/2., height/2,
             f'{value:.3f}', ha='center', va='center',
             fontsize=11, fontweight='bold', color=color)

# MAE Comparison
bars2 = ax2.bar(models, mae_values, color=colors_mae, alpha=0.9, width=0.6)
ax2.set_title('Mean Absolute Error (MAE)', fontsize=14, fontweight='bold', pad=25, color='#2C3E50')
ax2.set_ylabel('Score', fontsize=12, color='#2C3E50')

# Add value labels on bars
for i, (bar, value) in enumerate(zip(bars2, mae_values)):
    height = bar.get_height()
    color = 'white' if i == 1 else '#2C3E50'  # White text on red bar, dark on blue bars
    ax2.text(bar.get_x() + bar.get_width()/2., height/2,
             f'{value:.3f}', ha='center', va='center',
             fontsize=11, fontweight='bold', color=color)

# Style both axes
for ax in [ax1, ax2]:
    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#BDC3C7')
    ax.spines['bottom'].set_color('#BDC3C7')

    # Set tick parameters
    ax.tick_params(axis='both', colors='#2C3E50', labelsize=10)
    ax.tick_params(axis='x', rotation=0)

    # Add subtle grid
    ax.grid(True, alpha=0.2, axis='y', color='#BDC3C7')
    ax.set_axisbelow(True)

# Add red dots at the bottom to connect with highlighted bars (like your reference)
for ax in [ax1, ax2]:
    y_min = ax.get_ylim()[0]
    for i, model in enumerate(models):
        if i == 1:  # PatchTST (best performer)
            ax.plot(i, y_min - (ax.get_ylim()[1] - ax.get_ylim()[0]) * 0.05,
                   'o', color=highlight_color, markersize=8, clip_on=False)

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)

# Add overall title
fig.suptitle('Time Series Models Performance Benchmark',
             fontsize=16, fontweight='bold', y=0.95, color='#2C3E50')

# Show the plot
plt.show()

print("\n" + "="*60)
print("MODEL PERFORMANCE SUMMARY")
print("="*60)
print(f" BEST PERFORMER: PatchTST")
print(f"   MSE: {mse_values[1]:.4f} (Lowerst)")
print(f"   MAE: {mae_values[1]:.4f} (2nd Lowest)")
print("\nOther Models:")
for i, model in enumerate(models):
    if i != 1:  # Skip PatchTST as it's already highlighted
        print(f"   {model}: MSE={mse_values[i]:.4f}, MAE={mae_values[i]:.4f}")
print("="*60)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Model performance data
models = ['TCN', 'PatchTST', 'ARIMA', 'LLMPatchTST']
mse_values = [1.434791088104248, 1.368680477142334, 1.7947, 1.1298346636339363]
mae_values = [0.7879840731620789, 0.8820255994796753, 0.7842, 0.7263656142872726]

# Color scheme - highlight best performer (PatchTST) in red, others in blue tone
base_color = '#4E95D9'  # Your specified blue
highlight_color = '#E74C3C'  # Red for best performer

# Best performer is PatchTST (index 1 - lowest MSE and 2nd lowest MAE)
colors_mse = [base_color if i != 3 else highlight_color for i in range(len(models))]
colors_mae = [base_color if i != 3 else highlight_color for i in range(len(models))]

# Function to create and style a chart
def create_chart(metric_values, title, colors):
    fig, ax = plt.subplots(figsize=(8, 6))
    fig.patch.set_facecolor('white')

    # Create bars
    bars = ax.bar(models, metric_values, color=colors, alpha=0.9, width=0.6)

    # Add title and labels
    ax.set_title(title, fontsize=14, fontweight='bold', pad=25, color='#2C3E50')
    ax.set_ylabel('Score', fontsize=12, color='#2C3E50')

    # Add value labels on bars
    for i, (bar, value) in enumerate(zip(bars, metric_values)):
        height = bar.get_height()
        color = 'white' if i == 1 else '#2C3E50'  # White text on red bar, dark on blue bars
        ax.text(bar.get_x() + bar.get_width()/2., height/2,
                f'{value:.3f}', ha='center', va='center',
                fontsize=11, fontweight='bold', color=color)

    # Style the axes
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_color('#BDC3C7')
    ax.spines['bottom'].set_color('#BDC3C7')

    # Set tick parameters
    ax.tick_params(axis='both', colors='#2C3E50', labelsize=10)
    ax.tick_params(axis='x', rotation=0)

    # Add subtle grid
    ax.grid(True, alpha=0.2, axis='y', color='#BDC3C7')
    ax.set_axisbelow(True)

    # Add red dots at the bottom for best performer
    y_min = ax.get_ylim()[0]
    for i, model in enumerate(models):
        if i == 3:  # LLMPatchTST (best performer)
            ax.plot(i, y_min - (ax.get_ylim()[1] - ax.get_ylim()[0]) * 0.05,
                   'o', color=highlight_color, markersize=8, clip_on=False)

    plt.tight_layout()
    return fig

# Create MSE Chart
mse_fig = create_chart(mse_values, 'Mean Squared Error (MSE)', colors_mse)
plt.show()
# save the figures
mse_fig.savefig('mse_comparison.png', bbox_inches='tight', dpi=300)


# Create MAE Chart
mae_fig = create_chart(mae_values, 'Mean Absolute Error (MAE)', colors_mae)
plt.show()
# save the figures
mae_fig.savefig('mae_comparison.png', bbox_inches='tight', dpi=300)


# Print performance summary
print("\n" + "="*60)
print("MODEL PERFORMANCE SUMMARY")
print("="*60)
print(f" BEST PERFORMER: PatchTST")
print(f"   MSE: {mse_values[1]:.4f} (Lowerst)")
print(f"   MAE: {mae_values[1]:.4f} (2nd Lowest)")
print("\nOther Models:")
for i, model in enumerate(models):
    if i != 1:  # Skip PatchTST as it's already highlighted
        print(f"   {model}: MSE={mse_values[i]:.4f}, MAE={mae_values[i]:.4f}")
print("="*60)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
import matplotlib.patches as mpatches

# Set style for publication-quality plots
plt.style.use('default')
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['font.size'] = 12

# Model performance data
models = ['TCN', 'PatchTST', 'ARIMA', 'LLMPatchTST']
mse_values = [1.434791088104248, 1.368680477142334, 1.7947, 1.1298346636339363]
mae_values = [0.7879840731620789, 0.8820255994796753, 0.7842, 0.7263656142872726]

# Enhanced color palette - sophisticated and professional
colors = {
    'primary': '#2E86AB',      # Deep blue
    'secondary': '#A23B72',    # Deep magenta
    'tertiary': '#F18F01',     # Orange
    'highlight': '#C73E1D',    # Deep red - best performer
    'background': '#F8F9FA',   # Light gray background
    'text': '#2F3E46',         # Dark gray text
    'grid': '#E9ECEF'          # Light grid
}

# Identify best performer (lowest values for both metrics)
best_performer_idx = 3  # LLMPatchTST has lowest MSE and MAE

# Color assignment
bar_colors = [colors['primary'], colors['secondary'], colors['tertiary'], colors['highlight']]
bar_colors[best_performer_idx] = colors['highlight']

def create_aesthetic_chart():
    # Create figure with custom size and layout
    fig = plt.figure(figsize=(16, 10))
    fig.patch.set_facecolor('white')

    # Create subplots with custom spacing
    gs = fig.add_gridspec(2, 2, height_ratios=[1, 1], width_ratios=[1, 1],
                         hspace=0.3, wspace=0.25, left=0.08, right=0.95,
                         top=0.88, bottom=0.12)

    # MSE subplot
    ax1 = fig.add_subplot(gs[0, 0])
    # MAE subplot
    ax2 = fig.add_subplot(gs[0, 1])
    # Combined comparison subplot
    ax3 = fig.add_subplot(gs[1, :])

    # Helper function to style individual axes
    def style_axis(ax, title, ylabel):
        ax.set_title(title, fontsize=16, fontweight='bold', pad=20, color=colors['text'])
        ax.set_ylabel(ylabel, fontsize=13, fontweight='medium', color=colors['text'])
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_color(colors['grid'])
        ax.spines['bottom'].set_color(colors['grid'])
        ax.tick_params(axis='both', colors=colors['text'], labelsize=11)
        ax.grid(True, alpha=0.3, axis='y', color=colors['grid'], linestyle='-', linewidth=0.5)
        ax.set_axisbelow(True)
        ax.set_facecolor('white')

    # MSE Chart
    bars1 = ax1.bar(models, mse_values, color=bar_colors, alpha=0.85, width=0.7,
                    edgecolor='white', linewidth=2)
    style_axis(ax1, 'Mean Squared Error (MSE)', 'MSE Score')

    # Add value labels on MSE bars
    for i, (bar, value) in enumerate(zip(bars1, mse_values)):
        height = bar.get_height()
        text_color = 'white' if i == best_performer_idx else colors['text']
        ax1.text(bar.get_x() + bar.get_width()/2., height/2,
                f'{value:.3f}', ha='center', va='center',
                fontsize=11, fontweight='bold', color=text_color)

    # MAE Chart
    bars2 = ax2.bar(models, mae_values, color=bar_colors, alpha=0.85, width=0.7,
                    edgecolor='white', linewidth=2)
    style_axis(ax2, 'Mean Absolute Error (MAE)', 'MAE Score')

    # Add value labels on MAE bars
    for i, (bar, value) in enumerate(zip(bars2, mae_values)):
        height = bar.get_height()
        text_color = 'white' if i == best_performer_idx else colors['text']
        ax2.text(bar.get_x() + bar.get_width()/2., height/2,
                f'{value:.3f}', ha='center', va='center',
                fontsize=11, fontweight='bold', color=text_color)

    # Combined comparison chart (grouped bar chart)
    x = np.arange(len(models))
    width = 0.35

    # Normalize values for better visual comparison (optional)
    bars3_mse = ax3.bar(x - width/2, mse_values, width, label='MSE',
                       color=bar_colors, alpha=0.85, edgecolor='white', linewidth=1.5)
    bars3_mae = ax3.bar(x + width/2, mae_values, width, label='MAE',
                       color=bar_colors, alpha=0.6, edgecolor='white', linewidth=1.5)

    style_axis(ax3, 'Model Performance Comparison', 'Error Score')
    ax3.set_xlabel('Models', fontsize=13, fontweight='medium', color=colors['text'])
    ax3.set_xticks(x)
    ax3.set_xticklabels(models)

    # Enhanced legend
    legend = ax3.legend(loc='upper right', frameon=True, fancybox=True, shadow=True,
                       fontsize=12, title='Metrics', title_fontsize=13)
    legend.get_frame().set_facecolor('white')
    legend.get_frame().set_alpha(0.9)

    # Add value labels on combined chart
    for i, (bar_mse, bar_mae, mse_val, mae_val) in enumerate(zip(bars3_mse, bars3_mae, mse_values, mae_values)):
        text_color = 'white' if i == best_performer_idx else colors['text']

        # MSE labels
        ax3.text(bar_mse.get_x() + bar_mse.get_width()/2., bar_mse.get_height()/2,
                f'{mse_val:.3f}', ha='center', va='center',
                fontsize=10, fontweight='bold', color=text_color, rotation=0)

        # MAE labels
        ax3.text(bar_mae.get_x() + bar_mae.get_width()/2., bar_mae.get_height()/2,
                f'{mae_val:.3f}', ha='center', va='center',
                fontsize=10, fontweight='bold', color=text_color, rotation=0)

    # Add crown emoji or star for best performer
    crown_props = dict(boxstyle="round,pad=0.3", facecolor=colors['highlight'], alpha=0.8)
    ax3.text(best_performer_idx, max(max(mse_values), max(mae_values)) * 1.1, ' BEST',
             ha='center', va='bottom', fontsize=12, fontweight='bold',
             color='white', bbox=crown_props)

    # Main title
    fig.suptitle('Time Series Forecasting Model Performance Benchmark',
                fontsize=20, fontweight='bold', color=colors['text'], y=0.95)

    # Add performance summary text box
    summary_text = f"""Performance Summary:
 Best Performer: {models[best_performer_idx]}
    MSE: {mse_values[best_performer_idx]:.4f} (Lowest)
    MAE: {mae_values[best_performer_idx]:.4f} (Lowest)

 Ranking by MSE: {models[best_performer_idx]} < PatchTST < TCN < ARIMA
 Ranking by MAE: {models[best_performer_idx]} < ARIMA < TCN < PatchTST"""

    # Add text box with summary
    props = dict(boxstyle='round,pad=0.8', facecolor=colors['background'], alpha=0.9, edgecolor=colors['grid'])
    fig.text(0.02, 0.02, summary_text, fontsize=11, verticalalignment='bottom',
             bbox=props, color=colors['text'], fontfamily='monospace')

    # Add subtle watermark/signature
    fig.text(0.98, 0.02, 'Generated with Python & Matplotlib',
             fontsize=9, alpha=0.6, ha='right', va='bottom',
             style='italic', color=colors['text'])

    return fig

# Create the aesthetic chart
fig = create_aesthetic_chart()

# Display the plot
plt.show()

# Save with high quality
fig.savefig('model_performance_benchmark.png', dpi=300, bbox_inches='tight',
            facecolor='white', edgecolor='none', format='png')
fig.savefig('model_performance_benchmark.pdf', bbox_inches='tight',
            facecolor='white', edgecolor='none', format='pdf')

# Print detailed performance analysis
print("\n" + "="*70)
print(" COMPREHENSIVE MODEL PERFORMANCE ANALYSIS")
print("="*70)
print(f" CHAMPION: {models[3]} (LLMPatchTST)")
print(f"   MSE: {mse_values[3]:.6f}  Lowest (Best)")
print(f"   MAE: {mae_values[3]:.6f}  Lowest (Best)")
print(f"   Performance Advantage: {((mse_values[1]/mse_values[3])-1)*100:.1f}% better than 2nd place")

print(f"\n COMPLETE RANKINGS:")
print("   MSE Ranking (Lower is Better):")
mse_ranking = sorted(enumerate(mse_values), key=lambda x: x[1])
for rank, (idx, score) in enumerate(mse_ranking, 1):
    marker = "" if rank == 1 else "" if rank == 2 else "" if rank == 3 else "  "
    print(f"     {marker} {rank}. {models[idx]}: {score:.6f}")

print("\n   MAE Ranking (Lower is Better):")
mae_ranking = sorted(enumerate(mae_values), key=lambda x: x[1])
for rank, (idx, score) in enumerate(mae_ranking, 1):
    marker = "" if rank == 1 else "" if rank == 2 else "" if rank == 3 else "  "
    print(f"     {marker} {rank}. {models[idx]}: {score:.6f}")

print("="*70)
print(" Recommendation: Deploy LLMPatchTST for optimal forecasting performance")
print("="*70)
```
